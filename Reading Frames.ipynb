### Lendo o txt com a sequência de DNA

```python
with open ("rosalind_orf-4.txt", "r") as data:
    data = data.readlines()
        
DNA = ""
for line in data:
    if line[0] != ">":
        DNA += line.strip()

print (DNA)
```

### Criando um dicionário para o quadro de leitura (trinca de bases --> aminoácido)

```python
table = "UUU F      CUU L      AUU I      GUU V UUC F      CUC L      AUC I      GUC V UUA L      CUA L      AUA I      GUA V UUG L      CUG L      AUG M      GUG V UCU S      CCU P      ACU T      GCU A UCC S      CCC P      ACC T      GCC A UCA S      CCA P      ACA T      GCA A UCG S      CCG P      ACG T      GCG A UAU Y      CAU H      AAU N      GAU D UAC Y      CAC H      AAC N      GAC D UAA Stop   CAA Q      AAA K      GAA E UAG Stop   CAG Q      AAG K      GAG E UGU C      CGU R      AGU S      GGU G UGC C      CGC R      AGC S      GGC G UGA Stop   CGA R      AGA R      GGA G UGG W      CGG R      AGG R      GGG G"
table = table.split()
    
aa_dic = {}
for i in range (0, (len(table)), 2):
    aa_dic[table[i]] = table[i+1]
    
print (aa_dic)
```

### Obtendo a fita complementar de um sequencia de DNA

```python
comp = DNA.replace("A", "t").replace("T", "A").replace("t", "T").replace("C", "g"). replace("G", "C"). replace("g", "G")

comp = comp[::-1]

print (comp)
```

### Convertendo as duas fitas de DNA para RNA

```python
RNA = DNA.replace("T", "U")
comp = comp.replace("T", "U")

print (RNA)
print (comp)
```

### Obtendo todos os possíveis quadros de leitura

```python
read_1 = RNA
read_2 = read_1[1:]
read_3 = read_2[1:]

read_4 = comp
read_5 = read_4[1:]
read_6 = read_5[1:]

reads = []
reads.append(read_1)
reads.append(read_2)
reads.append(read_3)
reads.append(read_4)
reads.append(read_5)
reads.append(read_6)

for i in range (1, len(reads)+1):
    print ("Read", i, ":", reads[i-1])
```

### Obtendo os códons, traduzindo e detectando o início e fim de cada proteína

```python
prots = []
for read in reads:
    codons = []
    codon = ""
    for i in range (0, len(read), 3):
        codon += read[i : i+3]
        if len(codon) == 3:
            codons.append(codon)
            codon = ""
           
    print ("\nSequencia de codons:", codons)
    
    prot = ""
    for codon in codons:
        prot += aa_dic[codon]
    prots.append(prot)
        
for protein in prots:
    print ("\nSequencia traduzida:", protein)             
```

### Encontrando o códon de início e fim para cada sequência

```python
results = []

for protein in prots:
    protein = protein.replace("Stop", "sStop").split("Stop")
    for frag in protein:
        if "M" in frag and "s" in frag:
            frag = frag.replace("s", "")
            for w in range (len(frag)):
                if frag[w] == "M" and frag[w:] not in results:
                    results.append(frag[w:])
                    
for result in results:
    print (result)
```
